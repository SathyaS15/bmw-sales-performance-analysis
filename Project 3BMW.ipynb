Final Insights:

1.Top Performing Model: Certain BMW models contribute the highest share of total sales revenue, indicating strong market demand and customer preference.

2️.Price vs Sales Volume: Higher-priced models generally show lower sales volume compared to mid-range models, reflecting price sensitivity in the market.

3️.Regional Sales Distribution: Sales performance varies significantly across regions, with metropolitan areas generating higher revenue.

4️. Monthly/Yearly Trend Analysis: Sales show seasonal or yearly growth patterns, indicating demand fluctuations over time.

5️. Revenue Concentration: A small number of models contribute a large percentage of total revenue (Pareto principle effect).

6️. Engine Type Preference: Petrol/Diesel/Hybrid variants show different demand levels, highlighting customer fuel preference trends.

7️. Luxury Segment Stability: Premium BMW variants maintain steady demand despite higher pricing.

8️. Sales Growth Pattern: Certain years or months show strong growth rates, indicating successful marketing or economic recovery periods.

9️. Average Selling Price (ASP) Insight: The overall average selling price helps understand BMW’s positioning in the luxury automobile market.

10.Customer Demand Trend: Mid-segment luxury models attract a wider customer base compared to top-end flagship models.
